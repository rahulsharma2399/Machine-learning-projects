```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import statsmodels.api as smf
```

```python
data=pd.read_csv(r"C:\Users\VAIO\Desktop\Dataset 2\50_Startups.csv")
```

```python
data.head()

```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,4].values

```

```python
le=LabelEncoder()
```

```python
x[:,3]=le.fit_transform(x[:,3]) #converting the state column value in 0,1,2
```

```python
ohe=OneHotEncoder(categorical_features=[3])
```

```python
x=ohe.fit_transform(x).toarray()
#check x we have add multiple cloumns for each state
```

```python
x=x[:,1:] # to avoid dummy variable trap
x#we have 5 colmun
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=0)
mlr=LinearRegression()
mlr.fit(train_x,train_y)
```

```python
pred=mlr.predict(test_x)
```
